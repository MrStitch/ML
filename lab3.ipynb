```python
import math
import numpy
from matplotlib import pyplot as plt
from scipy.optimize import minimize

#Function F(x) 
def f(x):
    return math.sin(x / 5) * math.exp(x / 10) + 5 * math.exp(-x / 2)

#Gradient

res = minimize(f, 2, method='BFGS') #optimize for 2
print(round(f(float(res.x)), 2))
res  = minimize(f, 30, method='BFGS')#optimize for 30
print(round(f(float(res.x)), 2))



y = numpy.arange(1, 30, 0.5)
plt.plot(y, list(map(f, y)))
```
