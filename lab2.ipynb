```python
import re
import numpy
import math
from matplotlib import pyplot as plt

# функция
def f(x):
    return math.sin(x / 5) * math.exp(x / 10) + 5 * math.exp(-x / 2)
#создали матрицу NxN
def createMatrix(x):
    n = len(x)
    matrix = numpy.zeros(shape=(n, n))
    for i in range(n):
        for j in range(n):
            matrix[i][j] = x[i] ** j
    return matrix
# Создаем массив точек (f (x [i])
def getSolutions(rank, scope):
    y = numpy.zeros(shape=(len(scope)))
    for i in range(len(scope)):
        for j in range(len(rank)):
                y[i] += rank[j] * (scope[i] ** j) 
    return y

def linearEquationSolver(x): #функция линейных систем
    m = createMatrix(x)
    v = list(map(f, x))
    return numpy.linalg.solve(m, v) 

scope = numpy.arange(1, 15, .1) 
x = [[1, 15], [1, 8, 15], [1, 4, 10, 15]]
buffer = [] 
y = []

for i in range(3):
    buffer.append(linearEquationSolver(x[i])) 
                    
yMain = list(map(f, scope))
for i in range(len(buffer)):
    y.append(getSolutions(buffer[i], scope)) 

for i in range(len(y)): # рисую
    plt.plot(scope, y[i])
plt.plot(scope, yMain)
plt.show()
```
